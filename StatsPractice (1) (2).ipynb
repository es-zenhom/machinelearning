## Exercise 5: Rescaling a feature 

Rescaling is a common preprocessing task in machine learning. Many of the algorithms will assume all features are on the same scale, typically 0 to 1 or -1 to 1. There are a number of rescaling techniques, but one of the simlest is called min-max scaling. Min-max scaling uses the minimum and maximum values of a feature to rescale values to within a range. See https://en.wikipedia.org/wiki/Feature_scaling#Rescaling_(min-max_normalization)

## Exercise 6: Standardizing a feature

A common alternative to min-max scaling is rescaling of features to be approximately standard normally distributed. To achieve this, we use standardization to tranform the data such that it has a mean of 0 and a standard deviation of 1. Specifically, each element in the feature is transformed so that:
$$
x'_i = \frac{x_i-\bar x}{\sigma}
$$
where $\bar x$ and $\sigma$ are respectively the mean and the standard deviation of the original feature $x$. After the transformation, $x'$ has mean of zero: $\bar x' = 0$.

Create an array to work with:

```
# create a feature
feature = np.array([
    [-1000.1],
    [-200.2],
    [500.5],
    [600.6],
    [9000.9]
])
```

```
feature
```

Now standardize this feature

1. Using the formula above in numpy
2. The "standard scaler"
3. The "robust scaler"

Check whether the transformed feature has mean 0 and standard deviation 1.

```
feautre_1=(feature-np.average(feature))/np.std(feature)
print(feautre_1)
print('mean is: ',np.average(feautre_1), 'std is: ', np.std(feautre_1))
```

```
from sklearn.preprocessing import StandardScaler

ss = StandardScaler().fit(feature)

print(ss.transform(feature))
print('mean is: ',np.average(ss.transform(feature)), 'std is: ', np.std(ss.transform(feature)))
```

```
from sklearn.preprocessing import RobustScaler
rs = RobustScaler().fit(feature)
print(rs.transform(feature))
print('mean is: ',np.average(rs.transform(feature)), 'std is: ', np.std(rs.transform(feature)))
```

## Exercise 7: Generating Polynomial and Interaction Features

Polynomial features are needed when we suspect non-linear dependence on the feature variables. For example, if there are two features $x_1$ and $x_2$, by using the feature matrix as is, we will typically be including only a linear dependence on $x_1$ and $x_2$:
$$
y(x_1,x_2) \sim constant + a_1 x_1 + a_2 x_2
$$
What if we also want to include non-linear terms like $x_1^2$ and $x_2^2$ and the "interaction term" $x_1x_2$? The answer: just add those terms to the feature matrix. Here is an example:


```
# load libraries
from sklearn.preprocessing import PolynomialFeatures
```

```
# Create a simple 3x2 feature matrix to work with
features = np.array([[1,3],[2,4],[7,5]])
features
```

```
# Create PolynomialFeatures object
polynomial_interaction = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
```

```
# Create polynomial features
polynomial_interaction.fit_transform(features)
```

Now the columns represent the five features $x_1$, $x_2$, $x_1^2$, $x_1x_2$ and $x_2^2$.

Now create polynomial features for the same data:

1. Adding a constant term in addition to those above
2. Adding only nonlinear terms in which both $x_1$ and $x_2$ are present
3. Including all possible terms to 4th order.

```
polynomial_interaction_const = PolynomialFeatures(degree=2, interaction_only=False, include_bias=True)
polynomial_interaction_const.fit_transform(features)
```

```
print('number 2: ', np.delete(polynomial_interaction.fit_transform(features),3,1) )
```

```
polynomial_interaction_4th_order = PolynomialFeatures(degree=4, interaction_only=False, include_bias=False)
polynomial_interaction_4th_order.fit_transform(features)
```

## The nuclear option: delete every observation with missing values

Pandas dataframes provide an easy way to do this:

```
df_clean = df.dropna()
```

```
df_clean[0:30]
```

However, deleting observations can introduce bias into our data. There are three types of missing data:

1. Missing Completely At Random (MCAR). In this case, the probability that a value is missing is independent of everything. For example, a survey respondent rolls a die before answering each question. If she rolls a six, she skips that question and moves on to the next one.

2. Missing At Random (MAR). Here the probability that a value is missing is not completely random, but depends on the information captured in other features. In that case, removing the missing data is sort of OK because the other features still contain the information reflected in the presence of the missing data.

3. Missing Not At Random (MNAR). Here the probability that a value is missing is not completely random, and depends on information NOT captured in other features. In that case, removing the missing data injects bias because we are removing observations produced by some unobserved systematic effect.

## "Fudge" the data: replace the missing data with substitute values  

There are two main strategies here:

1. Use machine learning to predict the missing values. Treat the feature with missing values as a target vector and use the remaining subset of features to predict the missing values. Those types of algorithm will typically use the values at the nearest N observations as defined by some distance metric.

2. Fill in the missing values with some average (typical) values for the feature at hand.

With either of these methods, it is a good idea to create a binary variable which keeps track of which observations contain a modified (imputed) value.

## Exercise 1

Plot the functions $\sqrt{x}$, $x$ and $x^2$ in the interval $x\in (0,1)$ on the same plot. Use different line styles and include an appropriate legend, plot title and labels. Use equal aspect ratio on the x and y axes.

```
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import random
```

```
fig1 = plt.figure()
ax1 = plt.axes()
ax2 = plt.axes()
ax3 = plt.axes()

x = np.linspace(0, 1)
ax1.plot(x, np.sqrt(x),linestyle='solid',label='sqrt(x)');
ax2.plot(x, x,linestyle='dashed',label='x');
ax3.plot(x, x**2,linestyle='dotted',label='x^2');
plt.title("x, sqrt(x), and  x^2  curves");
plt.xlabel("x")
plt.legend();

```

## Exercise 2

Make a scatter plot showing two populations of 100 points each. Both populations are sampled from a two-dimensional Gaussian. For the first population, let the Gaussian be centered at (0,0) and have $\sigma_x=\sigma_y=1$. For the second population, let the Gaussian be centered at (2,2) and have $\sigma_x=\sigma_y=3$.

```
x=np.array([])
y=np.array([])

for i in range (100):
 x=np.append(x,random.gauss(0,1))
 y=np.append(y,random.gauss(0,1))
 
plt.plot(x, y,'o', color='black');


xx=np.array([])
yy=np.array([])

for i in range (100):
 xx=np.append(xx,random.gauss(2,3))
 yy=np.append(yy,random.gauss(2,3))

plt.plot(xx, yy,'o', color='blue');
plt.show()
```

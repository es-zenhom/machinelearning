## Exercise 6

Consider the Pauli matrices defined below.

$$
\left(
\begin{matrix}
0 & 1 \\
1 & 0
\end{matrix}
\right),
\left(
\begin{matrix}
0 & -i \\
i & 0
\end{matrix}
\right),
\left(
\begin{matrix}
1 & 0 \\
0 & -1
\end{matrix}
\right)
$$

1. Verify that they are traceless
2. Verify that they all have determinants -1.
3. Verify that they are Hermitian.
4. Verify that they are unitary.
5. Verify the SU(2) algebra, i.e., the commutation relations among the Pauli matrices. If you don't remember the formulas, derive them.


````
sigX = np.matrix([[0+0j,1+0j],[1+0j,0+0j]])
sigY = np.matrix([[0+0j,0-1j],[0+1j,0+0j]])
sigZ = np.matrix([[1+0j,0+0j],[0+0j,-1+0j]])
print(sigX) 
print(sigY)
print(sigZ)
````

````
print(np.trace(sigX))
print(np.trace(sigY))
print(np.trace(sigZ))
````

````
print(np.linalg.det(sigX))
print(np.linalg.det(sigY))
print(np.linalg.det(sigZ))
````

now let's get the hermitian of it's matrix (conjecgate trsnpose and they are gonna look the same as the orgingal matrix meaning that pauli matrices are hermitan)

````
print(sigX.getH)
print(sigY.getH)
print(sigZ.getH)
````

to check they are unitary (if we multiply the the matrix by it's conjegat transpose we must get identity matrix) since they are hermitian(the inverse transpose is the same as the matrix itsel so it's enough to get the square of every matrix and check that it's gonna equal identity)

````
print(np.dot(sigX,sigX))
print(np.dot(sigY,sigY))
print(np.dot(sigZ,sigZ))
print(np.identity(2)==np.dot(sigX,sigX))
print(np.identity(2)==np.dot(sigY,sigY))
print(np.identity(2)==np.dot(sigZ,sigZ))
````

now we check the SU2 comutation relation which is cyclic relation with the result being the missing matrix multiplied by complex (2i) (we cycle clockwise (xy then yz then zx)

````
print(np.dot(sigX,sigY)-np.dot(sigY,sigX)==2*(0+1j)*sigZ)
print(np.dot(sigY,sigZ)-np.dot(sigZ,sigY)==2*(0+1j)*sigX)
print(np.dot(sigZ,sigX)-np.dot(sigX,sigZ)==2*(0+1j)*sigY)
````
